# Basic Tensor Operations

## Normal Matrix Addition

```python
A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

B = np.array([
    [10, 20, 30],
    [40, 50, 60]
])

# Normal Matrix Addition
C = A + B

print(C)
```

## Broadcasting Addition

### Case 1: Add scalar

```python
# Scalar is added to every element

C = A + 5
print(C)
```

### Case 2: Add row vector

```python
# Row vector is broadcasted across rows

v = np.array([10, 20, 30])
C = A + v
print(C)
```

### Case 3: Add column vector

```python
# Column vector is broadcasted across columns

v = np.array([[10],
              [20]])

C = A + v
print(C)
```

## Matrix Multiplication

$$ C_{ij} = \sum_k A_{ik} B_{kj}$$

```python
import numpy as np

A = np.array([
    [1, 2],
    [3, 4]
])

B = np.array([
    [5, 6],
    [7, 8]
])

C = A @ B   # or np.matmul(A, B)

print(C)
```

## Hadamard Product (Element-wise)

$$ C_{ij} = A_{ij} B_{ij} $$

```python
# Multiply corresponding elements
# Same shape required

C = A * B

print(C)
```

## Transpose

Swap rows and columns.

$$ A_{ij}^T = A_{ji} $$

### Transpose of a Matrix

```python
A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

AT = A.T

print(AT)
```

### Transpose of a Vector

```python
v = np.array([1, 2, 3])

print(v.T)
```

👉 No change!

Because this is a 1D array, not a column vector.

```python
v = np.array([[1],
              [2],
              [3]])

print(v.T)
```

### Transpose using function

```python
np.transpose(v)
```

## Dot Product of Vectors

$$\vec{a} \cdot \vec{b}= \mathbf{a}^T \mathbf{b} = \mathbf{b}^T \mathbf{a}$$ 

```python
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

dot = np.dot(v1, v2)

print(dot)
```

# Norm

Norm is a way of measuring the "length" of vectors, matrices etc. Mathematically, a norm is any function $f$ that satisfies

- $ f(x) = 0 \implies x = 0 $
- $ f(x + y) \leq f(x) + f(y) $
- $ \forall  \alpha \in 	\mathbb{R}, f(\alpha x) = |\alpha|f(x) $

## Vector Norms

### $p$-norm / $L^p$-norm

$p$-norm / $L^p$-norm of a vector $\mathbf{𝑣} \in 	\mathbb{R}^n$

$$\left\| v \right\|_{p} = \left( |v_1|^p + |v_2|^p + |v_3|^p + \cdots + |v_n|^p \right)^{\frac{1}{p}}, \quad \text{for } p \geq 1 $$

$$\left\| v \right\|_{p} = \left( \sum_{i=1}^{n} |v_i|^p\right)^{\frac{1}{p}}, \quad \text{for } p \geq 1 $$

### $1$-norm / $L^1$-norm / Manhattan norm

$p=1$ → Manhattan norm

$$ \left\| v \right\|_{1} = |v_1| + |v_2| + |v_3| + \cdots + |v_n| $$ 

### $2$-norm / $L^2$-norm / Euclidean norm

$p=2$ → Euclidean norm (most common)

$$ \left\| v \right\|_2 = (v_1^2 + v_2^2 + v_3^2 + \cdots + v_n^2)^{\frac{1}{2}}$$

### Max norm

$p \to \infty$ → Max norm

$$\|v\|_{\infty} = \max_i |v_i| $$

```python
import numpy as np

# Sample vector
v = np.array([3, -4, 2])

# ---------------------------
# L1 Norm (Manhattan)
# ---------------------------
l1 = np.sum(np.abs(v))

# ---------------------------
# L2 Norm (Euclidean)
# ---------------------------
l2 = np.sqrt(np.sum(v**2))

# ---------------------------
# General Lp Norm
# ---------------------------
def lp_norm(v, p):
    return np.sum(np.abs(v)**p) ** (1/p)

l3 = lp_norm(v, 3)   # example for p=3

# ---------------------------
# L-infinity Norm (Max norm)
# ---------------------------
linf = np.max(np.abs(v))

# ---------------------------
# Using built-in NumPy function
# ---------------------------
l1_np = np.linalg.norm(v, ord=1)
l2_np = np.linalg.norm(v, ord=2)
l3_np = np.linalg.norm(v, ord=3)
linf_np = np.linalg.norm(v, ord=np.inf)

# ---------------------------
# Print results
# ---------------------------
print("Vector:", v)
print("L1 norm:", l1)
print("L2 norm:", l2)
print("L3 norm:", l3)
print("L-infinity norm:", linf)

print("\nUsing np.linalg.norm:")
print("L1:", l1_np)
print("L2:", l2_np)
print("L3:", l3_np)
print("L-infinity:", linf_np)
```

## Matrix Norms

### Frobenius Norm

$$
\|A\|_F = \sqrt{\sum_{i=1}^{m} \sum_{j=1}^{n} |a_{ij}|^2}
$$

```python
# Example matrix
A = np.array([
    [1, -2, 3],
    [4, 0, -6]
])

# ---------------------------
# Frobenius norm
# ---------------------------
fro = np.sqrt(np.sum(A**2))

# ---------------------------
# Using built-in NumPy
# ---------------------------
fro_np = np.linalg.norm(A, ord='fro')

print("Frobenius:", fro)
print("Frobenius:", fro_np)
```

# Linear Combination of Vectors

A linear combination of the set of vectors $\{\mathbf{v_1}, \mathbf{v_2}, \cdots \mathbf{v_n}\}$ is given by multiplying each vector by a corresponding scalar coefficient and adding the results.

$$ \sum_{i=1}^n \alpha_i \mathbf{v_i}$$

# Matrix Myltiplication as a Linear Combination of Column Vectors

Matrix mutiplication can be interpreted in terms of the linear combination of the columns.

```python
V = np.array([[1, 2], [2, 0], [3, 3]])
A = np.array([[1, 3], [2, 4]])
```

```python
V @ A
```

# Span of a set of Vectors

The span of a set of vectors is the set of all the vectors obtainable by a linear combination of the original vectors.

Ex: The span of vectors $ \mathbf{v_1} = (1, 0)$, $\mathbf{v_2} = (0, 1)$ is whole of $\mathbb{R}^2$.

# Column Space of a Matrix

The span of all the columns of a matrix is called the column space.

Note: The equation $\mathbf{A} \mathbf{x} = \mathbf{b}$ has a solution only if $\mathbf{b}$ lies in the column space of $\mathbf{A}$.

# Linear Independence

A set of vectors are linearly independent if none of these vectors can be written as a linear combination of the other vectors.

Ex: 
- $ \mathbf{v_1} = (1, 0)$, $\mathbf{v_2} = (0, 1)$  are linearly independent.
- $ \mathbf{v_1} = (1, 0)$, $\mathbf{v_2} = (0, 1)$, $\mathbf{v_3} = (3, 4)$  are linearly dependent.

Mathematically, $\mathbf{S} = {\mathbf{v_1}, \mathbf{v_2}, \cdots, \mathbf{v_n}} $ are linearly independent if and only if the linear combination $\alpha_1 v_1 + \alpha_2 v_2 + \cdots + \alpha_n v_n = 0 $ means that all the $\alpha_i = 0$.
