# Back Propagation

A neural network has layers:

<div align="center">Input &rarr; Hidden Layers &rarr; Output</div>

Each layer:

$$\mathbf{z^l} = \mathbf{W^l} \mathbf{a^{l-1}} + \mathbf{b}^l$$

$$\mathbf{a}^l = g(\mathbf{z^l}) $$

Where:
- $\mathbf{a}^{l-1}$ : input to layer
- $\mathbf{W}^l$ : weights
- $\mathbf{b^l}$ : bias
- $\mathbf{z^l}$ : linear output (weighted inputs)
- $\mathbf{a}^l$ : activation

Updating parameters / weights:
$$\mathbf{W} = \mathbf{W} - \eta \nabla L$$

Where
- $L$ : loss

## Numerical Gradient (Finite Differences) for finding $\nabla L$

Forward difference approximation:

$$ \frac{\partial L}{\partial w_1} \approx \frac {L(w_1 + \Delta w_1) - L(w_1)}{\Delta w_1}$$

Central difference (more accurate):

$$ \frac{\partial L}{\partial w_1} \approx \frac {L(w_1 + \epsilon) - L(w_1 - \epsilon)}{2 \epsilon}$$

### Problem with Finite Differences method

Too slow + computationally impossible for neural networks.

#### Problem 1: Too Many Parameters

A neural network can have:
- 1 million weights (very small model)
- 100 million+ weights (normal deep learning)

With this method, for each weight, we must:
- Change $w_i$
- Run full forward pass
- Compute loss

So, total cost:
<div align="center"> Cost = $O(n)$ forward passes </div>

If $𝑛 = 10^6$ weights:
- 1 forward pass = 10 ms
- Time taken per update =  10,000 seconds (~3 hours)


Backprop computes all gradients at once:

<div align="center"> Cost = $O(1)$ forward + backward pass </div>

If $𝑛 = 10^6$ weights:
- 1 forward pass = 10 ms
- 1 backward pass = 10 ms
- Time taken per update using backprop = $\sim$ 20 ms

#### Problem 2: Redundant Computation

In this method:
- Every time we tweak a weight, we recompute entire network again.
- But most computations are repeated.

Backprop solves this by:
- Reusing intermediate results (dynamic programming).

#### Problem 3: Numerical Instability

Your method depends on small $\epsilon$
- Too small &rarr; floating point errors
- Too large &rarr; inaccurate gradient

Example issue:

$$ L(w + \epsilon) \approx L(w)$$

Hence, subtraction becomes unstable (loss of precision).

#### Problem 4: Doesn’t Scale with Depth

Deep networks:
- Have many layers
- Strong interdependencies

Numerical gradient:
- Treats each parameter independently
- Doesn’t exploit structure

Backprop:
- Uses chain rule across layers efficiently

### Why Backprop Wins

Backprop uses:
- Chain Rule + Reuse of Computation

Instead of:
$$ \text{compute gradient per weight separately} $$

It does:

$$ \text{compute gradient for ALL weights in one pass}$$
